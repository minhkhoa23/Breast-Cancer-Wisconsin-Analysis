# Phân tích dữ liệu và áp dụng một vài kiến thức cơ bản về machine learning

## Các thư viện cần thiết

```python
import numpy as np
import pandas as pd
from ucimlrepo import fetch_ucirepo #load dataset
import matplotlib.pyplot as plt #ve bieu do
import seaborn as sns #ve bieu do
```

## Phân tích dữ liệu

```python
# fetch dataset 
breast_cancer_wisconsin_original = fetch_ucirepo(id=15) 
  
# data (as pandas dataframes) 
X = breast_cancer_wisconsin_original.data.features 
y = breast_cancer_wisconsin_original.data.targets 
  
# metadata 
print(breast_cancer_wisconsin_original.metadata) 
  
# variable information 
print(breast_cancer_wisconsin_original.variables) 
```

## Vẽ các biểu đồ
### Các biểu đồ cột

```python
# Bieu do so luong target
plt.figure(figsize=(5,4))
y.value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.xticks(rotation=0)
plt.title("Phân phối nhãn (Class)")
plt.xlabel("Class (2=Benign, 4=Malignant)")
plt.ylabel("Số lượng")
plt.show()

# Bieu do cot so luong cua cac feature
features = X.columns

plt.figure(figsize=(16, 20))

for i, col in enumerate(features, 1):
    plt.subplot(5, 2, i)  # 5 hàng, 2 cột (vì có 9 feature)
    X[col].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Phân phối giá trị - {col}")
    plt.xlabel("Giá trị (1–10)")
    plt.ylabel("Số mẫu")

plt.tight_layout()
plt.show()

df = X.copy()
df["Class"] = y.values

plt.figure(figsize=(16, 20))

for i, col in enumerate(X.columns, 1):
    plt.subplot(5, 2, i)  # 5 hàng, 2 cột
    sns.boxplot(x="Class", y=col, data=df, palette="Set2")
    plt.title(f"Boxplot của {col} theo Class (2=Benign, 4=Malignant)")
    plt.xlabel("Class")
    plt.ylabel(col)

plt.tight_layout()
plt.show()
#hello world
```
